```python
path='Output/df_bars.shp'
df_bars.to_file(path)
```

```python
import pysal as ps
qW_bars=ps.queen_from_shapefile(path)
```

```python
# the weight matrix
Wmatrix, ids = qW_bars.full()
Wmatrix
```

### Spatial Lag

```python
Night_total=df_bars.Night_total
Night_totalLag=ps.lag_spatial(qW_bars, Night_total)
```

#### Create Quantiles for maping

```python
Night_totalLagQ10 = ps.Quantiles(Night_totalLag, k=10)
Night_totalLagQ10
```

#### Map the daily complaints lag

```python
f, ax = plt.subplots(figsize=(9, 9))
df_bars.assign(cl=Night_totalLagQ10.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Daily Bar Complaints Spatial Lag Deciles");
```

### Global Spatial Autocorrelation

```python
b,a = np.polyfit(Night_total, Night_totalLag, 1)
b,a 
```

### Moran Scatterplot

```python
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(Night_total, Night_totalLag, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(Night_total.mean(), Night_totalLag.min(), Night_totalLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(Night_totalLag.mean(), Night_total.min(), Night_total.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(Night_total, a + b*Night_total, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of Daily Bar Complaints')
plt.xlabel('Daily Bar Complaints');
```

#### Moran's I

```python
I_DailyComplaint = ps.Moran(Night_total, qW_bars)
```

```python
I_DailyComplaint.I, I_DailyComplaint.p_sim
```

### Local Spatial Autocorrelation



```python
LMo_Night_total = ps.Moran_Local(Night_total.values, qW_bars ,permutations=999)
```

```python
Night_total.shape, qW_bars.n
```

```python
LMo_Night_total.Is[0:5], LMo_Night_total.p_sim[0:5]
```

### LISA (Local Indicators of Spatial Autocorrelation)

```python
sigs = Night_total[LMo_Night_total.p_sim <= .001]
W_sigs = Night_totalLag[LMo_Night_total.p_sim <= .001]
insigs = Night_total[LMo_Night_total.p_sim > .001]
W_insigs = Night_totalLag[LMo_Night_total.p_sim > .001]
```

### Plot Signficant LISA

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(9, 9))
plt.plot(sigs, W_sigs, '.', color='firebrick')
plt.plot(insigs, W_insigs, '.k', alpha=.2)
plt.text(s='$I = %.3f$' % I_DailyComplaint.I, x=1, y=1, fontsize=18)

 # dashed vert at mean of the last year's PCI
plt.vlines(Night_total.mean(), Night_totalLag.min(), Night_totalLag.max(), linestyle='--')
 # dashed horizontal at mean of lagged PCI
plt.hlines(Night_totalLag.mean(), Night_total.min(), Night_total.max(), linestyle='--')

# red line of best fit using global I as slope
plt.plot(Night_total, a + b*Night_total, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of Daily Bar Complaints')
plt.xlabel('Daily Bar Complaints')
plt.show()
```
